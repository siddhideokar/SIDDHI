```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load CSV with proper encoding
df = pd.read_csv("Global YouTube Statistics.csv", encoding="ISO-8859-1")

# Check structure
df.info()
df.head()
```

```python
plt.figure(figsize=(15, 8))

#  Graph 1: Top 10 YouTube Channels by Subscribers
top10 = df.nlargest(10, 'subscribers')
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st plot
plt.barh(top10['Youtuber'], top10['subscribers'], color='purple')
plt.xlabel("Subscribers")
plt.title("Top 10 YouTube Channels by Subscribers")
plt.gca().invert_yaxis()
plt.grid(True)

#  Graph 2: Top 6 Categories as a Pie Chart
top_categories = df['category'].value_counts().head(6)
plt.subplot(2, 2, 2)  # 2nd plot in the 2x2 grid
plt.pie(top_categories.values, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 6 YouTube Channel Categories")
plt.axis('equal')

plt.tight_layout()
plt.show()


```

```python
# Sort the data by subscribers
df_sorted = df.sort_values(by='subscribers', ascending=False).reset_index()

# Create a figure with 1 row, 2 columns (side by side)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

#  Line Plot: Subscribers vs Rank
axs[0].plot(df_sorted.index + 1, df_sorted['subscribers'], color='blue', linewidth=2)
axs[0].set_title("Subscribers Trend by Channel Rank")
axs[0].set_xlabel("Channel Rank")
axs[0].set_ylabel("Subscribers")
axs[0].grid(True)

#  Histogram: Distribution of Subscribers
axs[1].hist(df['subscribers'], bins=30, color='orange', edgecolor='black')
axs[1].set_title("Distribution of Subscribers")
axs[1].set_xlabel("Subscribers")
axs[1].set_ylabel("Frequency")
axs[1].grid(True)

#  Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Group by category and get the max yearly earnings
category_salary = df.groupby('category')['highest_yearly_earnings'].max().sort_values(ascending=False)

# Plot
plt.figure(figsize=(11,6))
plt.bar(category_salary.index, category_salary.values, color='green')
plt.title("Highest Yearly Earnings by Channel Category")
plt.xlabel("Category")
plt.ylabel("Highest Yearly Earnings (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
